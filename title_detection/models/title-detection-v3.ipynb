```python
from ultralytics import YOLO

model = YOLO('yolov8l.pt') 
```

```python
model.train(data='/kaggle/input/dataset-title-detection-v3/data.yaml', epochs=300, imgsz=640, rect=True)
```

```python
results = model.val()

results_dict = results.results_dict

print("Precision:", results_dict['metrics/precision(B)'])
print("Recall:", results_dict['metrics/recall(B)'])
print("mAP@0.5:", results_dict['metrics/mAP50(B)'])
print("mAP@0.5-0.95:", results_dict['metrics/mAP50-95(B)'])
print("Fitness Score:", results_dict['fitness'])

print("Speed (Preprocess):", results.speed['preprocess'], "ms")
print("Speed (Inference):", results.speed['inference'], "ms")
print("Speed (Loss):", results.speed['loss'], "ms")
print("Speed (Postprocess):", results.speed['postprocess'], "ms")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = '/kaggle/working/runs/detect/train/results.csv'
data = pd.read_csv(csv_path)

# Vẽ đồ thị Loss
plt.figure(figsize=(10, 6))
plt.plot(data['epoch'], data['train/box_loss'], label='Train Box Loss')
plt.plot(data['epoch'], data['val/box_loss'], label='Validation Box Loss')
plt.plot(data['epoch'], data['train/cls_loss'], label='Train Classification Loss')
plt.plot(data['epoch'], data['val/cls_loss'], label='Validation Classification Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss over Epochs')
plt.legend()
plt.show()

# Biểu đồ Precision & Recall
plt.figure(figsize=(10, 6))
plt.plot(data['epoch'], data['metrics/precision(B)'], label='Train Precision')
plt.plot(data['epoch'], data['metrics/recall(B)'], label='Train Recall')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training Precision & Recall')
plt.legend()
plt.show()

# Biểu đồ mAP (mean Average Precision)
plt.figure(figsize=(10, 6))
plt.plot(data['epoch'], data['metrics/mAP50(B)'], label='mAP@0.5')
plt.plot(data['epoch'], data['metrics/mAP50-95(B)'], label='mAP@0.5-0.95')
plt.xlabel('Epoch')
plt.ylabel('mAP')
plt.title('mAP over Epochs')
plt.legend()
plt.show()
```

```python
# Test mô hình trên một ảnh cụ thể
import cv2 
import matplotlib.pyplot as plt

img_path = '/kaggle/input/image-title-demo/image-title-demo/12_1.png'
results = model(img_path)

result = results[0]

boxes = result.boxes

image = cv2.imread(img_path)

for i, box in enumerate(boxes):
    x1, y1, x2, y2 = box.xyxy[0].tolist()  
    x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
    
    color = (255, 0, 0)  
    thickness = 2  
    cv2.rectangle(image, (x1, y1), (x2, y2), color, thickness)

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.axis('off') 
plt.show()

for i, box in enumerate(boxes):
    print(f"Bounding box {i+1}: {box.xywh}")
```
